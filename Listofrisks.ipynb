```python
import ast
import pandas as pd 

df = pd.read_csv('flood_risk_data.csv')
df_small = df[['liste_libelle_risque','libelle_commune']]
```

```python
def str_to_list(cell):
    cell = ''.join(c for c in cell if c not in "'[]")
    cell = cell.split(', ')
    return cell

df_llr = pd.DataFrame(df['liste_libelle_risque'].apply(str_to_list))

df_small['liste_libelle_risque'] = df['liste_libelle_risque'].apply(str_to_list)

```
